```python

import os
import librosa
import json
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

```

```python
json_decode = json.load(open("rms_data.json"))
```

```python
json_decode=json.load(open("rms_data.json"))
FRAME_LENGTH = 1024
HOP_LENGTH = 512

frames = range(259) 
t = librosa.frames_to_time(frames, hop_length=HOP_LENGTH)

happy_zcr_mean = np.average(np.array(json_decode["rms"][:1000]), axis=0)
sad_zcr_mean   = np.average(np.array(json_decode["rms"][1000:2000]), axis=0)
energetic_zcr_mean = np.average(np.array(json_decode["rms"][2000:3000]), axis=0)
calm_zcr_mean    = np.average(np.array(json_decode["rms"][3000:4000]), axis=0)

plt.figure(figsize=(10, 10))
plt.plot(t, happy_zcr_mean, color="r")
plt.plot(t, sad_zcr_mean, color="b")
plt.plot(t, energetic_zcr_mean, color="y")
plt.plot(t, calm_zcr_mean, color="g")


plt.ylim(0.0, 0.25)
plt.title("Average RMS across segment")
red_patch = mpatches.Patch(color='red', label='happy')
blue_patch = mpatches.Patch(color='blue', label='sad')
yellow_patch = mpatches.Patch(color='yellow', label='excited')
green_patch = mpatches.Patch(color = 'green', label='calm')

plt.legend(handles=[red_patch, blue_patch, yellow_patch, green_patch])
plt.show()

print(happy_zcr_mean)
```
